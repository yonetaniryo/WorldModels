```python
import env
import matplotlib.pyplot as plt

plt.figure(figsize=[12, 12])
for i in range(1, 4):
    for j in range(1, 4):
        e = env.make_env_car_racing('car_racing_%d%d' % (i, j), seed=1)

        e.reset()
        for t in range(50):
            e.step([0, 1, 0])
        img = e.render(mode='rgb_array')
        p, _, _, _ = e.step([0, 1, 0])
        plt.subplot(3, 3, (i - 1) * 3 + j)
        plt.imshow(p)
```
